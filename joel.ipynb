# The Parks Metric

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('./pittsburgh_parks.csv')

# Count the number of parks in each neighborhood
park_counts = data['neighborhood'].value_counts()

# Get the top 10 neighborhoods with the most parks
top_parks_neighborhoods = park_counts.head(10)

# Plotting
plt.figure(figsize=(10, 6))
top_parks_neighborhoods.plot(kind='bar')
plt.title('Top 10 Neighborhoods with the Most Parks in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('./pittsburgh_parks.csv')

# List of top 10 neighborhoods
top_neighborhoods = [
    'East Liberty', 'Central Business District', 'Beechview', 'South Side Slopes',
    'Point Breeze', 'Squirrel Hill South', 'Mount Washington', 'Sheraden',
    'Hazelwood', 'South Side Flats'
]

# Filter the dataset to include only the top 10 neighborhoods
top_data = data[data['neighborhood'].isin(top_neighborhoods)]

# Group by neighborhood and type, then count the occurrences
grouped_data = top_data.groupby(['neighborhood', 'type']).size().unstack(fill_value=0)

# Plotting the data
grouped_data.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Quantity of Each Type of Park in Top 10 Neighborhoods in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.xticks(rotation=45)
plt.legend(title='Park Type')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('./pittsburgh_parks.csv')

# Define the top 10 neighborhoods
top_neighborhoods = [
    'East Liberty', 'Central Business District', 'Beechview', 'South Side Slopes',
    'Point Breeze', 'Squirrel Hill South', 'Mount Washington', 'Sheraden',
    'Hazelwood', 'South Side Flats'
]

# Filter the dataset to include only the top 10 neighborhoods
top_data = data[data['neighborhood'].isin(top_neighborhoods)]

# Group by neighborhood and type, then count the occurrences
grouped_data = top_data.groupby(['neighborhood', 'type']).size().reset_index(name='count')

# Pivot the data for the plot
pivot_data = grouped_data.pivot(index='neighborhood', columns='type', values='count').fillna(0)

# Plotting the data
pivot_data.plot(kind='bar', figsize=(14, 8), width=0.8)
plt.title('Count of Each Type of Park in Top 10 Neighborhoods in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.xticks(rotation=45)
plt.legend(title='Park Type')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('./pittsburgh_parks.csv')

# Filter the dataset to include only 'Open Space' and 'Park' types
filtered_data = data[data['type'].isin(['Open Space', 'Park'])]

# Group by neighborhood and sum the occurrences
grouped_data = filtered_data.groupby('neighborhood').size()

# Get the top 3 neighborhoods with the most 'Open Space' and 'Park' types
top_3_open_space_park = grouped_data.nlargest(3)

# Plotting the data
top_3_open_space_park.plot(kind='bar', figsize=(10, 6))
plt.title('Top 3 Neighborhoods with the Most Open Space and Park Type Parks in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.xticks(rotation=45)
plt.show()
```

Within this data set, we were able to split the parks into a handful of individual groups.

*Those include:*
- top neighborhoods with the most parks in Pittsburgh
- the counts of each type in each neighborhood
- determined the best types of parks for ducks (open areas & parks)
- top 3 neighborhoods with most amount of these specific parks
